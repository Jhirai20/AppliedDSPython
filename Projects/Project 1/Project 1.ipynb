## Generate a separate dataset for numerical and categorical variables

```python
#extract data set based on number value presence 
numeric_df= ds.select_dtypes(include=[np.number])
categoric_df=ds.select_dtypes(exclude=[np.number])

#generate a list of just the col
numericcol=numeric_df.columns.tolist()
categorycol=categoric_df.columns.tolist()
```

### Missing value treatment


###Remove the entire col since its mostly empty

```python
ds.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature','FireplaceQu'], axis=1, inplace=True)
ds.isna().sum(axis=0)
```

###Interpolate by Mean or Median or Mode 
Use Box plot and ditribution plot to decide which method to use

```python
sns.boxplot(ds.LotFrontage)
```

```python
sns.distplot(ds.LotFrontage)
```

```python
#Since this column is not skewed and shows a normal distribution I use Mean Imputation
ds['LotFrontage']= ds['LotFrontage'].fillna(ds['LotFrontage'].mean())
```

###Remove row with when less than 30% missing

```python
ds = ds.dropna(axis=0)
ds.isna().sum(axis=0)
```

### Identify the skewness and distribution

```python
numeric_df.skew(axis = 0, skipna = True)
```

```python
numeric_df.skew(axis = 0, skipna = True).plot()
plt.xticks(rotation=45)
```

### Pair plot for distribution and density

```python
warnings.filterwarnings("ignore")
for i in range(0,len(numericcol),2):
        plt.figure(figsize=(10,4))
        plt.subplot(121)
        sns.distplot(ds[numericcol[i]], hist=True, kde=True)
        plt.subplot(122)            
        sns.distplot(ds[numericcol[i+1]], hist=True, kde=True)
        plt.tight_layout()
        plt.show()
```

### Identify significant variables using a correlation matrix 

```python
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='viridis')
```

```python
#Top 10 Correlations W/ sale price
k = 10 #number of variables for heatmap
cols = numeric_df.corr().nlargest(k, 'SalePrice')['SalePrice'].index
cm = numeric_df[cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, cmap = 'viridis')
plt.xticks(rotation=45)
```

```python
# We see here the top 10 corrilations with sales price from left-highest to right-lowest.
#Overall Quality and general living space being the biggest factors on sales price.
```
